```python
# Ordinal Tanımlama

from pandas.api.types import CategoricalDtype
```

```python
df.cut.head()
```

```python
df.cut = df.cut.astype(CategoricalDtype(ordered = True)) # kategoriye çevirdik
```

```python
df.dtypes
```

```python
df.cut.head() # Görüldüğü gibi sıralama doğru değil
```

```python
cut_kategoriler = ["Fair","Good","Very Good","Premium","Ideal"]
```

```python
df.cut = df.cut.astype(CategoricalDtype(categories = cut_kategoriler,ordered = True))
```

```python
df.cut.head(1) # görüldüğü gibi önem sıralaması doğru
```
